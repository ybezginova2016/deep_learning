In this notebook, we will classify small images cifar10 dataset from tensorflow keras datasets. There are total 10 classes as shown below. We will use CNN for classification.

## Data Description: CIFAR-10 dataset

**The CIFAR-10 dataset consists of 60000 32x32 colour images in 10 classes, with 6000 per class. There are 50000 training images and 10000 test images.**

The dataset is divided into 5 training batches and one test batch, each with 10000 images. 

- The ***test*** batch contains exactly 1000 randomly-selected images from each other class. 

- The ***training*** baches contain the remaining images in random order, but some training batches may contain more images from one class from another.

#### Between them, the training batches contain exactly 5000 images from each class.



```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import warnings
warnings.simplefilter("ignore", category=RuntimeWarning)
pd.options.mode.chained_assignment = None 
```

# 1 Load the dataset

```python
(X_train, y_train), (X_test,y_test) = datasets.cifar10.load_data()
```

***y_train is a 2D array, for our classification having 1D array is good enough. So, we will convert this to now 1D array with a direct directory to the class. We are reshaping the y_train dataset.***

```python
y_train = y_train.reshape(-1,) 
# -1 means that we don't want to change that dimention
# (, ) blank second argument means we want to flatten the array into the list
y_train[:5]
```

```python
y_test = y_test.reshape(-1,)
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
def plot_sample(X, y, index):
    plt.figure(figsize=(15, 2))
    plt.imshow(X[index])
    plt.xlabel(classes[y[index]])
```

```python
plot_sample(X_train, y_train, 0);
```

```python
plot_sample(X_train, y_train, 1)
```

***Normalize the images to a number from 0 to 1. Image has 3 channels (R,G,B) and each value in the channel can range from 0 to 255. Hence to normalize in 0-->1 range, we need to divide it by 255.***

# 2 Normalizing the training data

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_train[0] / 255.0
```

# 3 Build simple artificial neural network (ANN) for image classification

Before building CNN (more complex) we will build a simple ANN first and see how the network works.

```python
ann = models.Sequential([
        layers.Flatten(input_shape=(32,32,3)),
        layers.Dense(3000, activation='relu'),
        layers.Dense(1000, activation='relu'),
        layers.Dense(10, activation='softmax')    
    ])

ann.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann.fit(X_train, y_train, epochs=5)
```

### ***Conclusion:*** using simple ANN, at the end of 5 epochs, accuracy is 49.5%.

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
y_pred = ann.predict(X_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```

# 4 Build a convolutional neural network (CNN) to train our images

```python
cnn = models.Sequential([
    
    # cnn
    # dense
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # dense
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax') # softmax is the most popular activation function for normalization
])
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, epochs=10)
```

### ***Conclusion:*** With CNN, at the end 5 epochs, accuracy was 76.7% which is a significant improvement over ANN. CNN's are best for image classification and gives superb accuracy. Also computation is much less compared to simple ANN as maxpooling reduces the image dimensions while still preserving the features.

```python
cnn.evaluate(X_test,y_test)
```
