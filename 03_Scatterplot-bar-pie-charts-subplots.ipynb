## Graph Plot

```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 2])
plt.show();
```

```python
x = [5, 8, 10]
y = [12, 16, 6]

plt.plot(x, y)
plt.title('info')
plt.ylabel('Y axis')
plt.ylabel('X axis')

plt.show();
```

## Adding style to the plot

```python
from matplotlib import style
```

```python
# show which style I am going to use
style.use('ggplot')
```

```python
x = [5, 8, 10]
y = [12, 16, 6]

x2 = [6, 9, 11]
y2 = [6, 15, 7]
```

```python
plt.plot(x, y, 'g', label='line one', linewidth=5)
plt.plot(x2, y2, 'c', label='line two', linewidth=5)
plt.title('Epic info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()

plt.grid(True, color = 'k') # color is black
plt.show();
```

```python
# %%
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
plt.plot(xs, ys)

# %%
plt.style.available

# %%
plt.style.use('seaborn-whitegrid')
plt.plot(xs, ys)

# %%
plt.style.use('fivethirtyeight')
plt.plot(xs, ys);
```

## Bar Graph

#### Bar graph is used to compare things between different groups. We are trying to measure the changes.

***In bar plot we use CATEGORICAL variables.***

```python
import matplotlib.pyplot as plt
```

```python
plt.bar([2,4,6,8,10],[8,6,2,5,6], label='Example two', color='g');
plt.legend()
plt.xlabel('Bar number')
plt.ylabel('Bar height')

plt.title('Info')

plt.show();
```
