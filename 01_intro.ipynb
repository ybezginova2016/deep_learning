## What is Data Science 

**Data Science** is a concept that unifies Statistics, Computer Science and Business Domain.

**Data science** uses specific methods, algorithms, and systems with an objective to derive predictions from data.

This can be any given domain like health care, finance, manufacturing, etc.

**Data Science** is focused on extracting knowledge, insights and trends from databases which are typically large.


## How to implement data science?
- You need to define the problem statement at first and also what you want to predict.
- You need to collect the data related to the platform and fine tune it to make it meaningful.
- You need to use a framework like Tensorflow, Azure ML, Keras, etc.
- You need to use an algorithm to apply on the data (***Training***)
- Finally, you should use some test data to validate accuracy of your training.
