```python
import cv2
import dlib
```

```python
# starting a video capture device (default webcam)
cap = cv2.VideoCapture(0)

# face detection
hog_face_detector = dlib.get_frontal_face_detector()

dlib_facelandmark = dlib.shape_predictor("/Users/yuliabezginova/PycharmProjects/00_files-for_NLP/shape_predictor_68_face_landmarks.dat")
```

```python
# infinite loop for reading the webcam capture 
# and using face shape predictor

while True:
    # reading the image of the webcam
    _, frame = cap.read()
    
    # converting the image into the gray scale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = hog_face_detector(gray)
    for face in faces:

        face_landmarks = dlib_facelandmark(gray, face)

        for n in range(0, 68):
            x = face_landmarks.part(n).x
            y = face_landmarks.part(n).y
            cv2.circle(frame, (x, y), 1, (0, 255, 255), 1)
    
    cv2.imshow("Face Landmarks", frame)

    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
